```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('kc_house_data.csv')
```

# How many houses have 1 bedroom?

```python
one_room = df.loc[(df['bedrooms'] == 1)].count()
more_room = df.loc[(df['bedrooms'] != 1)].count()

labels = ['One room', 'More than one room']
colors = ['r', 'b']

plt.pie([one_room['bedrooms'], more_room['bedrooms']], labels = labels, colors = colors)
plt.title('Houses that have only one bedroom compared to others')

plt.show()
```
