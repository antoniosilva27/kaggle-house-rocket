```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv('kc_house_data.csv')
```

# How many houses have 1 bedroom?

```python
one_room = df.loc[(df['bedrooms'] == 1)].count()
more_room = df.loc[(df['bedrooms'] != 1)].count()

plt.figure(figsize=(5,8))

labels = [f'One room: {one_room.bedrooms}', f'More than one room: {more_room.bedrooms}']
colors = ['y', 'b']
explode = (.4, .2)

plt.pie([one_room['bedrooms'], more_room['bedrooms']], labels = labels, colors = colors, explode=explode)
plt.title('Houses that have only one bedroom compared to others', fontsize=20)


plt.show()
```

# How many houses have more than 1 floor?

```python
one_floor = df.loc[(df['floors'] == 1)].count()
more_floor = df.loc[(df['floors'] != 1)].count()

plt.figure(figsize=(5,8))

labels = [f'One floor: {one_floor.floors}', f'More than one floor: {more_floor.floors}']
colors = ['b', 'orange']
explode = (.4, .2)

plt.pie([one_room['floors'], more_room['floors']], labels = labels, colors = colors, explode=explode)
plt.title('Houses that have only one floor compared to others', fontsize=20)

plt.show()
```

# How many houses are in good or bad condition

```python
condition_1 = df.loc[(df.condition) == 1].count()
condition_2 = df.loc[(df.condition) == 2].count()
condition_3 = df.loc[(df.condition) == 3].count()
condition_4 = df.loc[(df.condition) == 4].count()
condition_5 = df.loc[(df.condition) == 5].count()

plt.figure(figsize=(8,12))

conditions = [condition_1.condition, condition_2.condition, condition_3.condition, condition_4.condition, condition_5.condition]

labels = [f'Terrible: {condition_1.condition}', f'Bad: {condition_2.condition}', f'Medium: {condition_3.condition}', f'Great: {condition_4.condition}', f'Perfect: {condition_5.condition}']
explode = (.8,.0,.0,.0,.0)
plt.pie(conditions, labels = labels, explode=explode)

plt.title('House condition comparison', fontsize=20)

plt.show()
```
